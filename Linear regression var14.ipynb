# Dataset (var 14)

### 1. Dataset loading

```python
from sklearn import datasets

rand = random.randint(1,20)
print(rand)

data, target, coef = datasets.make_regression(n_samples = 140, n_features=14,n_informative =round(14/5),
                                             noise = rand,coef = True, random_state = 2)
```

### 2. Feature plots

```python
pyplot.figure(figsize(25, 25))

plot_number = 0

for feature_name in range(14):
    pyplot.subplot(7, 2, plot_number + 1)
    pyplot.scatter(list(map(lambda x : x[plot_number], data)), target, color = 'b')
    pyplot.title('feature ' + str(plot_number))
    pyplot.xlabel(str(plot_number))
    pyplot.ylabel('target')
    plot_number += 1
```

### 3. Linear regression model learning

```python
#http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression

from sklearn import linear_model

linear_regressor = linear_model.LinearRegression()
linear_regressor.fit(data, target)
predictions = linear_regressor.predict(data)
```

### 4. Learned regression model

```python
print ("Learned regression model")
print (
    "y = {:.3f}*x1 + {:.3f}*x2 + {:.3f}*x3 + {:.3f}*x4 + {:.3f}*x5 + {:.3f}*x6 + {:.3f}*x7 + {:.3f}*x8 + {:.3f}*x9  + {:.3f}*x10 + {:.3f}*x11+ {:.3f}*x12+ {:.3f}*x13 + {:.3f}*x14 +{:.3f}\n"
    .format(linear_regressor.coef_[0], linear_regressor.coef_[1],linear_regressor.coef_[2],
    linear_regressor.coef_[3],linear_regressor.coef_[4],linear_regressor.coef_[5],
    linear_regressor.coef_[6],linear_regressor.coef_[7],linear_regressor.coef_[8],
    linear_regressor.coef_[9],linear_regressor.coef_[10],linear_regressor.coef_[11],
    linear_regressor.coef_[12],linear_regressor.coef_[13],linear_regressor.intercept_))
```

```python
print ("True regression model")
print (
    "y = {:.3f}*x1 + {:.3f}*x2 + {:.3f}*x3 + {:.3f}*x4 + {:.3f}*x5 + {:.3f}*x6 + {:.3f}*x7 + {:.3f}*x8 + {:.3f}*x9  + {:.3f}*x10  + {:.3f}*x11  + {:.3f}*x12  + {:.3f}*x13  + {:.3f}*x14\n "
    .format(coef[0],coef[1],coef[2],coef[3],coef[4],coef[5],coef[6],coef[7],coef[8],
    coef[9],coef[10],coef[11],coef[12],coef[13]))
```

### 5. Basic Estimators

```python
from sklearn import metrics

mae = metrics.mean_absolute_error(target, predictions)
print ('MAE = ', mae)
mse = metrics.mean_squared_error(target, predictions)
print ('MSE = ', mse)
```
